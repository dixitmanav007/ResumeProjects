```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
from datetime import datetime
```

```python
customer=pd.read_csv('customers.csv')
customer
```

```python
products= pd.read_csv('products.csv')
products
```

```python
transactions= pd.read_csv('transactions.csv')
transactions
```

```python
orders=pd.read_csv("orders.csv")
orders
```

```python
orders.head()
```

## No of orders based on order status

```python
orders.groupby("order_status")["order_id"].count().plot.bar('TOTAL NO. OF ORDER BASED ON ORDER STATUS')
plt.title('TOTAL NO. OF ORDERS BASED ON ORDER STATUS')
plt.xlabel('order_status')
plt.ylabel('NO. OF ORDERS')
plt.grid(True)
```

### TOTAL NO OF ORDERS OF EACH YEAR

```python
orders["year"]=pd.to_datetime(orders["order_purchase_date"]).dt.year
```

```python
orders.groupby("year")["order_id"].count().plot.bar('TOTAL NO. OF ORDER OF EACH YEAR')
plt.title('TOTAL NO. OF ORDERS  OF EACH YEAR')
plt.xlabel('order_id')
plt.ylabel('NO. OF ORDERS')
plt.grid(True)
```

### TOTAL NO OF ORDERS OF EACH MONTH

```python
orders["month"]=pd.to_datetime(orders["order_purchase_date"]).dt.month
```

```python
orders.groupby("month")["order_id"].count().plot.bar('TOTAL NO. OF ORDER OF EACH MONTH')
plt.title('TOTAL NO. OF ORDERS  OF EACH MONTH')
plt.xlabel('order_id')
plt.ylabel('NO. OF ORDERS')
plt.grid(True)
```

### TOTAL NO OF ORDERS OF MONTH-YEAR WISE 

```python
orders["month_year"]=pd.to_datetime(orders["order_purchase_date"]).dt.to_period('M')
```

```python
orders.groupby("month_year")["order_id"].count().plot.bar('TOTAL NO. OF ORDER OF MONTH-YEAR WISE')
plt.title('TOTAL NO. OF ORDERS OF MONTH-YEAR WISE')
plt.xlabel('order_id')
plt.ylabel('NO. OF ORDERS')
plt.grid(True)
```

### TOTAL NO OF ORDERS EACH DAY OF WEEK 

```python
orders['week']=pd.to_datetime(orders['order_purchase_date']).dt.weekday
orders.groupby(['week'])['order_id'].count().plot.line('TOTAL NO. OF ORDERS EACH DAY OF WEEK')
plt.title('TOTAL NO. OF ORDERS EACH DAY OF WEEK')
plt.xlabel('WEEK NO')
plt.ylabel('NO. OF ORDERS')
```

###  Comparison between no of orders purchased monthlywise between 2017 and 2018 

```python
o1=orders[orders['year']==2017]
o2=orders[orders['year']==2018]
```

```python
o1.groupby(['month'])['order_id'].count().plot(kind='line',marker='o')
o2.groupby(['month'])['order_id'].count().plot(kind='line',marker='o')
plt.legend()
plt.title('Comparison between no of orders purchased monthlywise between 2017 and 2018')
plt.ylabel('No of orders')
```

### SALES TREND MONTH-YEAR WISE

```python
profit = pd.merge(orders,transactions, how='inner', on='order_id')
```

```python
profit.groupby("month_year")["sales_amt"].sum().plot(kind='line',marker='*')
plt.legend()
plt.title('TOTAL PROFIT MONTH-YEAR WISE')
plt.xlabel('Month Year')
plt.ylabel('sales_amt')
```

### TOTAL PROFIT MONTH-YEAR WISE

```python
profit = pd.merge(orders,transactions, how='inner', on='order_id')
```

```python
profit.groupby("month_year")["profit_amt"].sum().plot(kind='line',marker='o')
plt.legend()
plt.title('TOTAL PROFIT MONTH-YEAR WISE')
plt.xlabel('Month Year')
plt.ylabel('profit_amt')

```
