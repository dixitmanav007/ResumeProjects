### TOTAL NO OF ORDERS OF EACH YEAR

```python
orders['year']=pd.to_datetime(orders['order_purchase_date']).dt.year
orders.groupby(['year'])['order_id'].count()
```

### TOTAL NO OF ORDERS OF MONTH-YEAR WISE 

```python
orders['month_year']=pd.to_datetime(orders['order_purchase_date']).dt.to_period('M')
orders.groupby(['month_year'])['order_id'].count()   
```

### TOTAL NO OF ORDERS EACH DAY OF WEEK 

```python
orders['week']=pd.to_datetime(orders['order_purchase_date']).dt.weekday
orders.groupby(['week'])['order_id'].count()
```

### Difference between the delivered date and Estimate date

```python
orders['diff']=(pd.to_datetime(orders['order_estimated_delivery_date']))-(pd.to_datetime(orders['order_delivered_customer_date']))
orders[['order_id','order_estimated_delivery_date','order_delivered_customer_date','diff']]
```

### TOTAL SALES AMOUNT OF EACH CUSTOMER

```python
oc= pd.merge(orders, customers, on='customer_id', how='inner')
oct = pd.merge(oc, transactions, how='inner', on='order_id')
oct.groupby('customer_id')['sales_amt'].sum()
```

### TOTAL SALES AMOUNT MONTH-YEAR WISE

```python
oct['my1']=pd.to_datetime(oct['order_purchase_date']).dt.strftime('%m-%y')
print(oct['my1'])
```

### TOTAL PROFIT MONTH-YEAR WISE

```python
oct.groupby('my1')['profit_amt'].sum()
```

### TOTAL COUNT OF PRODUCT SALE OF EACH CATEGORY

```python
octp = pd.merge(oct, products, how='inner', on='product_id')
octp.head()
```
